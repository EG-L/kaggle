```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
url = 'https://book.naver.com/bestsell/bestseller_list.nhn'

driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver.implicitly_wait(30)
```

```python
driver.get(url)
bsO = BeautifulSoup(driver.page_source,'html.parser')
```

```python
book_url = []

for index in range(0,25):
  dl_data = bsO.find('dt',{'id':'book_title_'+str(index)})
  link = dl_data.select('a')[0].get('href')
  print(link)
  book_url.append(link)
```

```python
for index,book_page_url in enumerate(book_url):
  driver.get(book_page_url)
  bsO = BeautifulSoup(driver.page_source,'html.parser')

  title = bsO.find('meta',{'property':'og:title'}).get('content')
  author = bsO.find('dt',text='저자').find_next_siblings('dd')[0].text.strip()
  image = bsO.find('meta',{'property':'og:image'}).get('content')
  url = bsO.find('meta',{'property':'og:url'}).get('content')

  dd = bsO.find('dt',text='가격').find_next_siblings('dd')[0]
  salePrice = dd.select('div.lowest strong')[0].text
  originalPrice = dd.select('div.lowest span.price')[0].text
  disacount = dd.select('div.lowest span.discount')[0].text
  print(title, author, image, url, dd,salePrice,originalPrice,disacount)
```
